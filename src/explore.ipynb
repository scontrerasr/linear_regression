```python
# Your code here
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/linear-regression-project-tutorial/main/medical_insurance_cost.csv')
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.info()   
```

```python
for column in df.columns:
    if df[column].dtype not in ['int64', 'float64']:
        df[column] = pd.factorize(df[column])[0]
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in df.columns:
    if column not in ('charges'):
        fig, axis = plt.subplots(2, 1, figsize=(10, 7))
        sns.regplot(x=column, y='charges', data=df, ax=axis[0])
        sns.heatmap(df[[column,'charges']].corr(), annot=True, fmt=".2f", ax=axis[1], cbar=False)
        plt.tight_layout()
        plt.show()
```

```python
x=df[['smoker', 'age', 'bmi',]]
y=df['charges'] 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
X_train.head()
```

```python
for columna in X_train.columns:
    sns.histplot(X_train,x=columna)
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train,y_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns,index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns,index=X_test.index)

X_train_scaled.head()


```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test)

print("Intercept (a):", model.intercept_)
print("Coefficients (b1, b2, b3):", model.coef_)

```

```python
y_pred = model.predict(X_test_scaled)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R^2 Score:", r2_score(y_test, y_pred))
```
